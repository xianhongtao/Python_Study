1.定义一个函数，参数为一个实数，代表一个摄氏温度值（华氏温度值），将它转换为一个华氏温度值（摄氏温度值），并返回该值（F = 1.8C + 32或C=(F - 32) * 5/9）。在主程序中调用该函数。

```python
def tempreture_conversion(input_celsius):
    fahrenheit = (input_celsius * 9 / 5) + 32
    return fahrenheit


c_tempreture = 100
f_tempreture = tempreture_conversion(c_tempreture)
print(f"100摄氏度是{f_tempreture}华氏度")
```

2.定义一个函数，函数参数为一个小于10000的正整数，分解它的各位数字，并以一个元组的形式返回。在主程序中调用该函数。

```python
def digits_seperate(input):
    output = []
    while input != 0:
        output.append(input % 10)
        input = input // 10
    finaloutput = []
    for i in range(len(output) - 1, -1, -1):
        finaloutput.append(output[i])
    return finaloutput


print(digits_seperate(1234567890))
```

3.练习随机数应用。请生成30个随机数据，模拟一个班的考试成绩（要求在40～100分之间）。计算这批数据的平均分、最高分和最低分，并排序由高到低输出。（选做）

```python
import random


def gen_random_number_tulpe(count):
    output = []
    for i in range(count):
        output.append(random.randint(40, 100))
    return tuple(output)


score = gen_random_number_tulpe(30)
print(f"生成的成绩元组是{score}")
print(f"平均成绩是{sum(score) / len(score)}")
print(f"最高成绩是{max(score)}")
print(f"最低成绩是{min(score)}")
print(f"成绩从高到低排序是{sorted(score, reverse=True)}")
```
