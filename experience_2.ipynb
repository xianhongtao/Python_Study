1、生成一个长度为10的全0一维numpy数组，将索引为技术的元素设置为1，索引为偶数的元素设置为2。

```python
import numpy as np
arr = np.zeros(10)
for row_index in range(10):
    if row_index % 2 != 0:
        arr[row_index] = 1
    else:
        arr[row_index] = 2
print (arr)
```

2、生成一个4行4列的全0二维numpy数组，将0、2行和0、2列交叉点的元素设置为2。

```python
import numpy as np
arr = np.zeros((4, 4))
for row_index in range(4):
    for j in range(4):
        if row_index % 2 == 0 and j % 2 == 0:
            arr[row_index, j] = 2
print(arr)
```

3、给定一个一维NumPy数组arr，包含一系列整数。你需要编写一个Python程序来完成以下任务：计算数组arr中所有元素的平均值；找出数组arr中所有大于平均值的元素，并将这些元素乘以2；创建一个新的NumPy数组result，该数组包含原始数组arr中所有小于或等于平均值的元素，以及上一步中处理过的元素（按它们在原数组中的顺序）。输出result数组。

```python
import numpy as np
n = int(input("Enter a number: "))
arr = np.random.randint(0, 127, size=n)
print(arr)
result = []
sum = np.sum(arr)
avarage = sum / n
for row_index in range(n):
    if arr[row_index] > avarage:
        result.append(int(arr[row_index] * 2))
    else:
        result.append(int(arr[row_index]))
print(result)
```

4、编写一个程序，给定一个一维numpy数组arr，要求将数组中的最大值和最小值互换，生成一个新的数组并输出。

```python
import numpy as np
n = int(input("Enter a number: "))
arr = np.random.randint(0, 127, size=n)
print(arr)
numpy_arr_index_max = np.argmax(arr)
numpy_arr_index_min = np.argmin(arr)
arr[numpy_arr_index_max], arr[numpy_arr_index_min] = arr[numpy_arr_index_min], arr[numpy_arr_index_max]
print(arr)
```

5、编写一个程序，输入正整数m、n，创建一个m行n列的二维numpy随机数组，按行输出行索引不小于列索引的元素。

```python
import numpy as np
n = int(input("Enter n: "))
m = int(input("Enter m: "))
arr = np.random.randint(0, 127, size=(n, m))
result_arr = np.zeros((n, m))
print(arr)
for row_index in range(n):
    for j in range(m):
        if row_index >= j:
            result_arr[row_index, j] = arr[row_index, j]
        else:
            result_arr[row_index, j] = 0
print(result_arr)
```

6、编写一个程序，输入正整数m、n，创建一个m行n列的二维numpy随机数组，要求将矩阵的每一行按升序排列，生成一个新的矩阵并输出。

```python
import numpy as np
n = int(input("Enter n: "))
m = int(input("Enter m: "))
arr = np.random.randint(0, 127, size=(n, m))
print(arr)
for row_index in range(n):
    arr[row_index] = sorted(arr[row_index])
print(arr)
```

7、编写一个程序，输入正整数m、n，创建一个m行n列的二维numpy随机数组，其中包含了不同城市的温度数据（每行代表一天，每列代表一个城市）。编写一个Python程序来完成以下任务：计算每一天的平均温度（即每行的平均值）；找出平均温度最高的那一天，并输出那一天所有城市的温度。

```python
import numpy as np
total = 0
temp_highest = 0
n = int(input("Enter n: "))
m = int(input("Enter m: "))
arr = np.random.randint(-10, 45, size=(n, m))
print(arr)
for row_index in range(n):
    for j in range(m):
        total += arr[row_index, j]
    total = total / m
    print("第" + str(row_index) + "行的平均气温是" + str(total))
    if temp_highest < total:
        temp_highest = total
    total = 0
print("最高气温是" + str(temp_highest))
```
