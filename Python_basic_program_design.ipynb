1、编写一个 Python 程序，输入一组学生的成绩，存储在一个列表中。然后编写代码实现以下功能：计算并输出所有学生的平均成绩，输出高于平均分的学生成绩列表。

```python
StudentScoresList = []
StudentsAboveAverage = []
StudentNumber = int(input("请输入学生总数: "))

for i in range(StudentNumber):
    StudentScoresList.append(int(input("请输入学生成绩: ")))
print("列表内容为：", StudentScoresList)
StudentGradeAverage = sum(StudentScoresList) / StudentNumber
print("平均分为：", StudentGradeAverage)
for i in range(StudentNumber):
    if StudentScoresList[i] > StudentGradeAverage:
        StudentsAboveAverage.append(StudentScoresList[i])
print("高于平均分的学生成绩为：", StudentsAboveAverage)
```

2、某公司用一个字典记录一批商品的成本、售价和销量，字典项的格式为“商品编号:(成本价,售价,销量)”。编写程序计算该公司的盈利情况，并输出结果。

```python
NumberOfItems = int(input("请输入商品种类数: "))
ItemSellDict = {}

for i in range(NumberOfItems):
    ItemName = input("请输入商品名称: ")
    ItemPrice = float(input("请输入商品价格: "))
    ItemCost = float(input("请输入商品成本: "))
    ItemSellAmount = int(input("请输入商品售出数量: "))
    ItemSellDict[ItemName] = (ItemPrice, ItemCost, ItemSellAmount)

##字典结构：{商品名称：(价格，成本，售出数量)}
#print(ItemSellDict)

TotalIncome = 0
for Item in ItemSellDict:
    ItemPrice = ItemSellDict[Item][0]
    ItemCost = ItemSellDict[Item][1]
    ItemSellAmount = ItemSellDict[Item][2]
    TotalIncome += (ItemPrice - ItemCost) * ItemSellAmount
print("总收入为：", TotalIncome)
```

3、编写一个 Python 程序，用户输入一组字母组成的列表，然后通过选择结构判断并分类出元音字母和辅音字母，将它们分别存储在两个不同的列表中。程序最后输出元音字母和辅音字母的列表及其数量。

```python
CharacterList = list(input("请输入: ")) #不会有人输入一个列表的罢，还是转换一下吧

#print(CharacterList)

CharacterInVowelList = []
CharacterOffVowelList = []
VowelList = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']

for i in range(len(CharacterList)):
    if CharacterList[i] in VowelList:
        if CharacterList[i].isupper():
            CharacterList[i] = CharacterList[i].lower()
        CharacterInVowelList.append(CharacterList[i])
    else:
        if CharacterList[i].isupper():
            CharacterList[i] = CharacterList[i].lower()
        CharacterOffVowelList.append(CharacterList[i])

print("元音字母的列表为：", CharacterInVowelList, "数量是：", len(CharacterInVowelList))
print("非元音字母的列表为：", CharacterOffVowelList, "数量是：", len(CharacterOffVowelList))
```

4、随机生成一个长度为100的整数列表，列表的元素值在范围[100,1000]内，选出其中的素数并放到一个新列表中，输出新列表。

```python
import random, math

def PrimeNumberCheck(Number):
    import math
    IsPrime = True
    TensPlace = 1
    TensPlace = math.ceil(math.sqrt(Number)/10)
    if Number <= 1:
        IsPrime = False
    if Number % 2 == 0 or Number % 3 == 0 or Number % 5 == 0 or Number % 7 == 0:
        IsPrime = False
    for i in range(1 , TensPlace+1):
        if Number % (i*10+1) == 0 or Number % (i*10+3) == 0 or Number % (i*10+7) == 0 or Number % (i*10+9) == 0:
            IsPrime = False
    if Number in (2 , 3 , 5 , 7, 11 , 13 , 17 , 19):
        IsPrime = True
    return IsPrime

RandomNumberList = []
RandomPrimeNumberList = []
RandomGeneratorMinimum = 100
RandomGeneratorMaximum = 1000

for i in range(100):
    NewRandomNumber = random.randint(RandomGeneratorMinimum,RandomGeneratorMaximum)
    RandomNumberList.append(NewRandomNumber)

print("随机数列表为：", RandomNumberList)

for i in range(len(RandomNumberList)):
    if PrimeNumberCheck(RandomNumberList[i]):
        RandomPrimeNumberList.append(RandomNumberList[i])

print("随机数质数列表为：", RandomPrimeNumberList)
```

7、编写一个Python函数 find_max_min，该函数接收一个包含整数的列表，并返回该列表中的最大值和最小值。然后，编写代码调用该函数，输出结果。

```python
def find_max_min (List):
    return (max(List), min(List))

UserInput = input("请输入整数数列： ")
UserInputList = UserInput.split()
#print(UserInputList)

for i in range(len(UserInputList)):
    UserInputList[i] = int(UserInputList[i])
print(find_max_min(UserInputList))
```

8、假设你有一个包含多个整数的列表numbers，列表中的每个元素代表一个学生的分数。你需要编写一个Python程序，该程序遍历这个列表，找出所有分数在60分（包含60分）以上的学生，并将这些学生的分数存储在一个新的列表passing_scores中。同时，计算并输出这些通过考试的学生数量。

```python
numbers = []
StudentGradePassed = []

while True:
    try:
        number = int(input("输入学生成绩：")) 
        numbers.append(number)
    except:
        break

for i in range(len(numbers)):
    if numbers[i] > 60:
        StudentGradePassed.append(numbers[i])

print("及格学生成绩为：", StudentGradePassed)
```

10、一个二维列表保存了学生的实验成绩，成绩是整数，每一行的格式为“学号，姓名，实验一，实验二，实验三”。编写程序将该列表写入文本文件“chengji.txt”。

```python
import time
StudentGradeDict = [
    ("201901", "张三", 90, 80, 70),
    ("201902", "李四", 85, 75, 65),
    ("201903", "王五", 80, 70, 60)
]
with open("chengji.txt", "w", encoding='utf-8') as f:
    f.write(time.strftime("%Y-%m-%d %H:%M:%S", time.localtime()) + "\n")
    header = ["学号", "姓名", "实验一", "实验二", "实验三"]
    f.write("\t".join(header) + "\n")
    for Student in StudentGradeDict:
        f.write("\t".join(map(str, Student)) + "\n")
```
